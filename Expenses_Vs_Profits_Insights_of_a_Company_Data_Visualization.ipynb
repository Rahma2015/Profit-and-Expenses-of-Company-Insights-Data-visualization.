#It is a company data. It is expenses and profits at each year data. Some simple visualization was made on this data to know whether the company was making profits or making losses over consective 10  years

```python
import pandas as pd
data={
    "year":[2014,2015,2016,2017,2018,2019,2020,2021,2022,2023],
    "expenses":[120000,102000,90000,100000,120000,128000,110000,130000,150000,140000],
    "profits":[7000,9000,12000,10000,20000,18000,21000,25000,23000,27000]
}

dataframe=pd.DataFrame(data)
```

```python
dataframe
```

#I visualize year vs expenses

```python
import matplotlib.pyplot as plt
plt.plot(dataframe["year"],dataframe["expenses"])
plt.title("year vs expenses")
plt.xlabel('year')
plt.ylabel('expenses')
plt.show()
```

#I visualize year vs profits

```python
plt.plot(dataframe["year"],dataframe["profits"])
plt.title("year vs profits")
plt.xlabel('year')
plt.ylabel('profits')
plt.show()
```

#We compare between profits and expenses

---



#I used linear plotting:

```python
plt.plot(dataframe["year"],dataframe["expenses"],label="expenses")
plt.plot(dataframe["year"],dataframe["profits"],label="profits")
plt.legend(loc="upper right")
plt.title("expenses vs profits")
plt.xlabel('year',loc='right')
plt.show()
```

#Visulaization of year vs profit using bar chart

```python
plt.bar(dataframe["year"],dataframe["profits"])
plt.title("year vs profits")
plt.xlabel('year')
plt.ylabel("profits")
plt.show()
```

#Visulaization of year vs expenses using bar chart

```python
plt.bar(dataframe["year"],dataframe["expenses"])
plt.title("year vs expenses")
plt.xlabel('year')
plt.ylabel("expenses")
plt.show()
```

#Conclusion and Insights:

---



#We can say that over ten years starting from 2014 to 2023, the company pay more for expenses than when it gains from profits


---



#Also, we can say that using linear graph/plot is better than bar graph when comes to comparing between profits and expenses. It is easy to discover that expenses are higher when you use linear plot/graph.
